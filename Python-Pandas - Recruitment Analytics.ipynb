```python
# Filling missing candidate emails with Not Available

pipeline_df['candidate_email'] = pipeline_df['candidate_email'].fillna('Not Available')

# Filling Interview_Rounds with 0 for candidates who were not shortlisted for interviews

pipeline_df['interview_rounds'] = pipeline_df['interview_rounds'].fillna(0).astype(int)

# Replacing missing Backout_Reason values with 'No Backout' for candidates who did not withdraw or are still in process

pipeline_df['backout_reason'] = pipeline_df['backout_reason'].fillna('No Backout')

#Filling TAT_Days with median

pipeline_df['tat_days'] = np.where(
    pipeline_df['joining_status'] == 'Joined',
    (pipeline_df['joining_date'] - pipeline_df['submission_date']).dt.days,
    np.nan
)
```

```python
pipeline_df['tat_days'] = pipeline_df['tat_days'].clip(lower=10, upper=40)
```

```python
date_col = ['open_date', 'closing_date']

for col in date_col:
  jobs_df[col] = jobs_df[col].replace(r'^\s*$', None, regex=True)
  jobs_df[col] = pd.to_datetime(jobs_df[col], errors='coerce')
  jobs_df[col] = jobs_df[col].where(
      jobs_df[col].notna(),
      None
  )
```

```python
date_col=['submission_date', 'interview_date', 'offer_date', 'joining_date']

for col in date_col:
  pipeline_df[col] = pipeline_df[col].replace(r'^\s*$', None, regex=True)
  pipeline_df[col] = pd.to_datetime(pipeline_df[col], errors='coerce')
  pipeline_df[col] = pipeline_df[col].where(
      pipeline_df[col].notna(),
      None
  )
```

```python
date_col = ['start_date', 'end_date']

for col in date_col:
  finance_df[col] = finance_df[col].replace(r'^\s*$', None, regex=True)
  finance_df[col] = pd.to_datetime(finance_df[col], errors='coerce')
  finance_df[col] = finance_df[col].where(
      finance_df[col].notna(),
      None
  )
```

```python
#finance revenue marging from negative to 0

margin_cols = ['company_margin', 'total_margin']
finance_df[margin_cols] = finance_df[margin_cols].clip(lower=0)
```

```python
#Calculating the time taken to schedule interviews after candidate submission

pipeline_df['submission_to_interview_days'] = (pipeline_df['interview_date'] - pipeline_df['submission_date']).dt.days
```

```python
#Calculating the time taken to release offers after candidate interviews

pipeline_df['interview_to_offer_days'] = (pipeline_df['offer_date'] - pipeline_df['interview_date']).dt.days
```

```python
#Creating Offer_Acceptance_Flag to track candidates who accepted or declined job offers

pipeline_df['offer_acceptance'] = np.where(
    pipeline_df['joining_status'] == 'Joined',
    1,
    0
)
```

```python
# Categorizing candidates based on the number of days spent in the hiring pipeline

pipeline_df['tat_bucket'] = pd.cut(
    pipeline_df['tat_days'],
    bins=[0, 15, 30, 60, np.inf],
    labels=['Fast', 'Moderate', 'Slow', 'Critical']
)
```

```python
object_col = pipeline_df.select_dtypes(include='object').columns

for col in object_col:
  pipeline_df[col] = pipeline_df[col].replace(
      r'^\s*$', None, regex = True
  )

  pipeline_df[col] = pipeline_df[col].where(
      pipeline_df[col].notna(),
      None
  )
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(x=pipeline_df['expected_salary'])

plt.title("Expected Salary Outliers")
plt.show()
```

```python
# IQR Method

Q1 = pipeline_df['expected_salary'].quantile(0.25)
Q3 = pipeline_df['expected_salary'].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

pipeline_df = pipeline_df[
    (pipeline_df['expected_salary'] >= lower) &
    (pipeline_df['expected_salary'] <= upper)
]
```

```python
# EDA Recruiter Performance

plt.figure(figsize=(10,6))
pipeline_df.groupby('recruiter_name')['closure_flag'] \
    .sum() \
    .sort_values(ascending=False) \
    .head(10) \
    .plot(kind='bar')

plt.title("Top Recruiters by Closures")
plt.ylabel("Closures")
plt.show()
```
